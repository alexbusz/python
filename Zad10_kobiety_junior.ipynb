```python
dataset=pd.read_csv('dataset_pandas_pd.csv', delimiter=',', names=['imie', 'nazwisko', 'rok_urodzenia', 'miesiac_urodzenia', 'dzien_urodzenia', 
'miejsce_urodzenia', 'oddzial_miasto', 'departament', 'rok_zatrudnienia', 'miesiac_zatrudnienia', 'dzien_zatrudnienia',
'doswiadczenie(w_latach)'])
dataset
```

```python
dataset['poziom_doswiadczenia'] = ['Junior' if x<3 else 'Mid' if 4<=x<8 else 'Senior' if 9<=x<12 else 'Ekspert' for x in dataset['doswiadczenie(w_latach)']]
dataset
```

```python
dataset['plec'] = ['Kobieta' if x.endswith('a') else 'Mezczyzna' for x in dataset['imie']]
dataset
```

```python
data=dataset[(dataset['poziom_doswiadczenia']=='Senior')&(dataset['plec']=='Kobieta')].groupby('oddzial_miasto').oddzial_miasto.count()
data[data==data.max()]
data.idxmax()
```
